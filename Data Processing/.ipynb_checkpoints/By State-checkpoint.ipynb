```python
import pandas as pd
```

```python
# load data from 'Covid by State.csv'
covid = pd.read_csv('../Data/Covid by State.csv', usecols = ['state', 'fips', 'cases', 'deaths', 'confirmed_cases', 'confirmed_deaths'])

# create a calculated field 'death_rate_confirmed' = 'confirmed_deaths'/'confirmed_cases'
covid['death_rate_confirmed'] = covid['confirmed_deaths']/covid['confirmed_cases']

# create a calculated field 'death_rate' = 'deaths'/'cases'
covid['death_rate'] = covid['deaths']/covid['cases']

covid.head()
```

```python
# load data from 'Mask Use by State.csv'
maskUse = pd.read_csv('../Data/Mask Use by State.csv').rename(columns = {'State fips': 'fips'})
maskUse.head()
```

```python
# load data from 'Population by State.csv'
population = pd.read_csv('../Data/Population by State.csv').drop(['State'], axis = 1).rename(columns = {'State fips': 'fips'})
population.head()
```

```python
# load data from 'Hospital Utilization by State.csv'
hospital = pd.read_csv('../Data/Hospital Utilization by State.csv').drop(['OBJECTID', 'state_name'], axis = 1).rename(columns={'State_fips': 'fips'})
```

```python
# load data from 'Health Insurance by State.csv'
insurance = pd.read_csv('../Data/Health Insurance by State.csv').drop(['State'], axis = 1).rename(columns={'State_fips': 'fips'})
```

```python
# merge covid with maskUse, population, hospital, insurance
merged = pd.merge(covid, maskUse)
merged = pd.merge(merged, population)
merged = pd.merge(merged, hospital)
merged = pd.merge(merged, insurance)
```

```python
# create a calculated field 'cases/population' = 'cases'/Census'
merged['cases/population'] = merged['cases']/merged['Census']
merged.head()
```
