```python
import pandas as pd
```

```python
# load data from 'Covid by County.csv'
covid = pd.read_csv('../Data/Covid by County.csv')

# create a calculated field 'death_rate_confirmed' = 'confirmed_deaths'/'confirmed_cases'
covid['death_rate_confirmed'] = covid['confirmed_deaths']/covid['confirmed_cases']

# create a calculated field 'death_rate' = 'deaths'/'cases'
covid['death_rate'] = covid['deaths']/covid['cases']

covid.head()
```

```python
# load data from 'Mask Use by County.csv'
maskUse = pd.read_csv('../Data/Mask Use by County.csv')
maskUse.head()
```

```python
# load data from 'Population by County.csv'
population = pd.read_csv('../Data/Population by County.csv')
population.head()
```

```python
# merge covid with maskUse, population
merged = pd.merge(covid.drop('date', axis = 1), maskUse.rename(columns={'COUNTYFP': 'fips'}))
merged = pd.merge(merged, population.drop(['County', 'STATE'], axis = 1))
```

```python
# create a calculated field 'cases/population' = 'cases'/Census'
merged['cases/population'] = merged['cases']/merged['Census']
merged.head()
```

```python
hospital = pd.read_csv('../Data/Hospital Utilization by State.csv')
hospital.drop(['OBJECTID', 'state_name'], axis = 1).rename(columns={'State_fips': 'fips'})
```

```python
for i in ['fips', 'deaths']: merged[i] = merged[i].astype(int)
merged.to_csv('Aggregated by County.csv', index = False)
```
